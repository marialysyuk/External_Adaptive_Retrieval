```python
from multiel import BELA
import os
import pandas as pd
from tqdm import tqdm
import numpy as np
```

```python
#set intial threshold for NER and EL to 0.01 to increase amount of candidates
bela_run = BELA(
 md_threshold=0.1,
 el_threshold=0.1, 
 checkpoint_name ="wiki", 
#  device = "cuda:0",
 config_name="joint_el_mel_new",
 repo ="wannaphong/BELA"
)
```

### WDSQ

### 4 modes:
1) 'bela_base' -- take initial question and rub BELA pipeline
2) 'bela_large_case' -- take initial question, capitalize all words in the question and run BELA pipeline
3) 'bela_large_ner' -- take initial question, capitalize just NER (use the code below) and run BELA pipeline
4) 'bela_just_ner' -- take just NER from question (use the code below) and run BELA pipeline

```python
df_wdsq = pd.read_csv("AdaRAG/data/adaptive_rag_natural_questions/test.csv")
```

```python
for i in tqdm(range(len(df_wdsq))):
    df_wdsq.loc[i, 'bela_base'] = bela_run.process_batch([df_wdsq.loc[i, 'question_text']])
```

```python
#rerank bela entities by el scores
for k in tqdm(range(len(df_wdsq))):
    inds = np.argsort(df_wdsq.loc[k, 'bela_base'][0]['el_scores'])[::-1][:5]
    new_arr =[df_wdsq.loc[k, 'bela_base'][0]['entities'][i] for i in inds]
    offsets = [df_wdsq.loc[k, 'bela_base'][0]['offsets'][i] for i in inds]
    lengths = [df_wdsq.loc[k, 'bela_base'][0]['lengths'][i] for i in inds]
    cur_set = set()
    fin_ents = []
    fin_ents_lbls = []
    for j in range(len(new_arr)):
        if new_arr[j] not in cur_set:
            cur_set.add(new_arr[j])
            fin_ents.append(new_arr[j])
            start =  offsets[j]
            length = lengths[j]
            fin_ents_lbls.append(df_wdsq.loc[k, 'question_text'][start:start+length])
    df_wdsq.loc[k, 'bela_base_ents'] = ', '.join(fin_ents)
    df_wdsq.loc[k, 'bela_base_ents_lbls'] = ', '.join(fin_ents_lbls)
```
